```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data=data.drop(["RowNumber","CustomerId","Surname"],axis=1)
```

```python
data.head()
```

```python
label = LabelEncoder()
data["Gender"]= label.fit_transform(data["Gender"])
```

```python
data.tail()
```

```python
from sklearn.preprocessing import OneHotEncoder
oneHot = OneHotEncoder(sparse_output=False)
geo_data = oneHot.fit_transform(data[["Geography"]])
```

```python
oneHot.get_feature_names_out(["Geography"])
```

```python
geo_encoded_df=pd.DataFrame(geo_data,columns=oneHot.get_feature_names_out(["Geography"]))
```

```python
geo_encoded_df
```

```python
data = data.drop('Geography',axis =1)
```

```python
input_data = pd.concat([data,geo_encoded_df],axis=1)
```

```python
input_data
```

```python
x= input_data.drop("EstimatedSalary",axis =1)
y = input_data["EstimatedSalary"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
with open("scalar.pkl" ,"wb") as file:
    pickle.dump(scalar,file) 
```

```python
with open("label_encoder.pkl","wb") as file:
    pickle.dump(label,file)
```

```python
with open("oneHot.pkl","wb") as file:
    pickle.dump(oneHot,file)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(64,activation="relu",input_shape=(X_train.shape[1],)),
    Dense(32,activation="relu"),
    Dense(1)
])
```

```python
model.compile(optimizer="adam",loss="mean_absolute_error",metrics=["mae"])
```

```python
model.summary()
```

```python
log_dir = "reglogs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
TensorBoard_callBack = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
EarlyStopping_callback = EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
history = model.fit(
    X_train,y_train,
    validation_data=(X_test,y_test),
    epochs=100,
    callbacks=[EarlyStopping_callback,TensorBoard_callBack]
)
```

```python
%load_ext tensorboard
%tensorboard --logdir reglogs/fit
```

```python
loss, mae = model.evaluate(X_test, y_test)
print("Loss:", loss)
print("MAE:", mae)
```

```python
model.save("reg_model.h5")
```
